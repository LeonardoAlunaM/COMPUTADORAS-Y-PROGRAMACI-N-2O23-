```
import networkx as nx
import matplotlib.pyplot as plt

# Define los estados y los conjuntos A y B
estados = ["e0", "e1", "e2", "e3", "e4", "e5", "e6", "e7"]
conjunto_a = [1, 2, 5, 10, 20, 50, 100, 'V', 'N']
conjunto_b = [1, 2, 5, 10, 20, 50, 100, 'C', 'O']

# Inicializa el grafo
G = nx.DiGraph()

# Agrega nodos y aristas al grafo
G.add_nodes_from(estados)
for estado_inicial in estados:
    for simbolo in conjunto_a:
        estado_destino = "e" + str(conjunto_a.index(simbolo))
        G.add_edge(estado_inicial, estado_destino, label=str(simbolo))

# Dibuja el grafo
pos = nx.spring_layout(G, seed=42)  # Layout para dibujar el grafo
labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_nodes(G, pos, node_size=1000)
nx.draw_networkx_labels(G, pos, font_size=10)
nx.draw_networkx_edges(G, pos, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Diagrama de Transición de la MEF")
plt.axis('off')
plt.show()

# Tabla de Transición y Salida
tabla_transicion_salida = {}
for estado_inicial in estados:
    tabla_transicion_salida[estado_inicial] = {}
    for simbolo in conjunto_a:
        estado_destino = "e" + str(conjunto_a.index(simbolo))
        salida = conjunto_b[conjunto_a.index(simbolo)]
        tabla_transicion_salida[estado_inicial][str(simbolo)] = (estado_destino, salida)

print("\nTabla de Transición y Salida:")
for estado, transiciones in tabla_transicion_salida.items():
    print(f"{estado}:")
    for simbolo, (destino, salida) in transiciones.items():
        print(f"  {simbolo} -> {destino}, {salida}")

# Función de Transición y Función de Salida
def funcion_transicion(estado_actual, entrada):
    if entrada in conjunto_a:
        estado_destino = "e" + str(conjunto_a.index(entrada))
        return estado_destino
    else:
        return estado_actual

def funcion_salida(estado_actual, entrada):
    if entrada in conjunto_a:
        return conjunto_b[conjunto_a.index(entrada)]
    else:
        return None

# Validar con transiciones si la máquina recibe un billete de 100 y se oprime el botón verde.
estado_actual = "e0"
entrada = 100

while True:
    if estado_actual == "e0" and entrada == "V":
        print("\nLa máquina recibió un billete de 100 y se oprimió el botón verde.")
        break
    elif estado_actual == "e7":
        print("\nLa máquina ha llegado al estado final.")
        break
    estado_destino = funcion_transicion(estado_actual, entrada)
    salida = funcion_salida(estado_actual, entrada)
    print(f"Estado actual: {estado_actual}, Entrada: {entrada}, Estado destino: {estado_destino}, Salida: {salida}")
    estado_actual = estado_destino

plt.show()
```
