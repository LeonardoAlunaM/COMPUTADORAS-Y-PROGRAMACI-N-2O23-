```python
# CALCULADORA 1 UTILIZANDO LA LIBRERIA PYQT5 06 / 08 / 2023, 8:34 AM
import PyQt5.QtWidgets as qtw # Importando PyQt5

class MainWindow(qtw.QWidget): # Creando la clase MainWindow, la cual hereda qtw.QWidget, Esta clasa contendrá todos los elementos visuales
    def __init__(self):        # Además de ser a base en donde se colocarán todos los demás elementos, en la línea def __init__(self): se inicializan los elementos de la ventana
        super().__init__()
        self.setWindowTitle("CALCULADORA CON PYQT5")  # Agregando el título de la ventana
        self.setLayout(qtw.QVBoxLayout()) # Diseño para la ventana principal
        #btn1 = qtw.QPushButton('test') # "Test" es el pequeño texto que se muestra dentro del botón definido en la aplicación
        #self.layout().addWidget(btn1)  # Ahora se muestra el botón, e incluso se ilumina al colocar el mouse sobre él
        self.keypad()
        self.temp_nums = []  # Parte de la lógica que realmente se encarga de realizar los cálculos de la calculadora
        self.fin_nums = [] # También es parte de la lógica para realizar los cálculos de la calculadora

        self.show()

    def keypad(self):
        container = qtw.QWidget() # Dentro de esta línea se contendrán todos los elementos de la aplicación, mismos que ya se han definido arriba
        container.setLayout(qtw.QGridLayout())

        # Botones
        self.result_field = qtw.QLineEdit() # Muestra el texto ingresado por el usuario
        # btn_result = qtw.QPushButton('Ingresar', clicked = lambda: self.result_field.setText('test'))  # Lógica inicial para hacer funcionar los botones de la calculadora
        btn_result = qtw.QPushButton('Ingresar', clicked = self.func_result)  # El resultado y el botón borrar tienen sus propio método
        btn_clear = qtw.QPushButton('Limpiar', clicked = self.clear_calc )
        # Botones de los números
        btn_9 = qtw.QPushButton('9', clicked = lambda: self.num_press('9')) # Los botones tienen el mismo método, pero diferentes argumentos
        btn_8 = qtw.QPushButton('8', clicked = lambda: self.num_press('8'))
        btn_7 = qtw.QPushButton('7', clicked = lambda: self.num_press('7'))
        btn_6 = qtw.QPushButton('6', clicked = lambda: self.num_press('6'))
        btn_5 = qtw.QPushButton('5', clicked = lambda: self.num_press('5'))
        btn_4 = qtw.QPushButton('4', clicked = lambda: self.num_press('4'))
        btn_3 = qtw.QPushButton('3', clicked = lambda: self.num_press('3'))
        btn_2 = qtw.QPushButton('2', clicked = lambda: self.num_press('2'))
        btn_1 = qtw.QPushButton('1', clicked = lambda: self.num_press('1'))
        btn_0 = qtw.QPushButton('0', clicked = lambda: self.num_press('0'))
        # Botones para los operadores
        btn_suma = qtw.QPushButton('+', clicked = lambda: self.func_press('+'))
        btn_resta = qtw.QPushButton('-', clicked = lambda: self.func_press('-'))
        btn_multiplicar = qtw.QPushButton('*', clicked = lambda: self.func_press('*'))
        btn_dividir = qtw.QPushButton('%', clicked = lambda: self.func_press('%'))
        # btn_modulo = qtw.QPushButton('9')  Agregar más botones eventualmente, este servirá para devolver el módulo de una operación

        # Agregando los botones al diseño, el cual es una cuadrícula
        container.layout().addWidget(self.result_field,0,0,1,4) # Este campo ocupa 4 columnas dentro del diseño cuadriculado
        container.layout().addWidget(btn_result,1,0,1,2)
        container.layout().addWidget(btn_clear,1,2,1,2)
        container.layout().addWidget(btn_9,2,0)
        container.layout().addWidget(btn_8,2,1)
        container.layout().addWidget(btn_7,2,2)
        container.layout().addWidget(btn_suma,2,3)
        container.layout().addWidget(btn_6,3,0)
        container.layout().addWidget(btn_5,3,1)
        container.layout().addWidget(btn_4,3,2)
        container.layout().addWidget(btn_resta,3,3)
        container.layout().addWidget(btn_3,4,0)
        container.layout().addWidget(btn_2,4,1)
        container.layout().addWidget(btn_1,4,2)
        container.layout().addWidget(btn_multiplicar,4,3)
        container.layout().addWidget(btn_0,5,0,1,3)
        container.layout().addWidget(btn_dividir,5,3)
        self.layout().addWidget(container) # Agregando realmente los botones a la calculadora, de momento los votones no tienen ninguna funcionalidad implementada

    def num_press(self, key_number):
        self.temp_nums.append(key_number)
        temp_string = ''.join(self.temp_nums)
        if self.fin_nums:
            self.result_field.setText(''.join(self.fin_nums) + temp_string)
        else:
            self.result_field.setText(temp_string)

    def func_press(self,operator):
        temp_string = ''.join(self.temp_nums)
        self.fin_nums.append(temp_string)
        self.fin_nums.append(operator)
        self.temp_nums = []
        self.result_field.setText(''.join(self.fin_nums))

    def func_result(self):  # Está función se utiliza para calcular el campo del resultado de las operaciones
        fin_string = ''.join(self.fin_nums) + ''.join(self.temp_nums)
        result_string = eval(fin_string)
        fin_string += '='
        fin_string += str(result_string)
        self.result_field.setText(fin_string)

    def clear_calc(self):  # Vacia el campo de resultados de la calculadora para los siguientes cálculos
        self.result_field.clear()
        self.temp_nums = []
        self.fin_nums =  []


app = qtw.QApplication([]) # Se crea el seguimiento de todas las acciones que la aplicación llevará por debajo de la interfaz gráfica
mw = MainWindow()

app.setStyle(qtw.QStyleFactory.create('Fusion'))


app.exec_() # Le índica a Python el hecho de que se tiene que ejecutar la aplicación
# Con las líneas 1 a 12 ya se muestra una ventana en la pantalla al momento de ejecutar el código, esto sin contar la línea que específica el
# nombre de la ventana y algunas otras que continue agregando después de las primeras 8 que efectivamente intervienen en el funcionamiento más
# básico de la presente aplicación

# Según parece la calculadora ya funciona, el problema es que de momento no distinge la jerarquía de las operaciones y además de eso
# Al momento de ingresar los datos directamente desde el teclado, así como otros símbolos, el programa de la caculadora se cae
# La única manera en la cual el programa puede ser medianamente efectivo de momento, es efectuar las operaciones paso por paso
# Es decir ingresar los datos para una operación, anotar el resultado, y en seguida realizar los calculos posteiores
```
